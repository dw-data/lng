```python
import pandas as pd
import re
```

## Data Cleaning / Converting

```python
energymix = pd.read_csv('../data/export-selected_countries_dataset.csv')
energymix.head()
```

```python
energymix.info()
```

```python
#to do: check value_counts of exceptions

# replacing non-numerical values with 0 for aggregation

#N/A - Not Available.  "N/A" means that data is expected, not received and therefore missing.
#N/E - not expected. "N/E" means that data is not expected, data provider is not configured for this data item,   area, border.
#"-" - ahead of time. " - " means that data is ahead of time and therefore can't be published at the moment. This usually can be seen for future time intervals.
#"  " - blank space. Blank space is published when data doesn't exist. This can happen for example for balancing data (17.1.f), when there are no activated offers and therefore price don't exists.

columns = energymix.columns.to_list()

for column in columns:
    if column != "Area" and column != "MTU":
        energymix[column] = [str(value).replace("n/a", "0").replace("n/e", "0").replace(" ", "0").replace("-","0") for value in energymix[column]]
        energymix[column] = energymix[column].astype(float)
    else:
        pass
```

```python
energymix.info()
```

```python
#reformatting date
energymix['date_parsed']=[re.sub('(\d\d\.\d\d\.\d\d\d\d)(.*)', '\g<1>', value) for value in energymix['MTU']]
```

```python
#creating a key from country-day to groupby and use aggregation method
energymix['key']=energymix['Area']+energymix['date_parsed']
```

```python
# Gruppieren und aggregieren
grouped_energymix = energymix.groupby(by='key').agg('sum').reset_index()
grouped_energymix
```

```python
# extracting country in new column
grouped_energymix['country'] = [re.sub('(.*)\([A-Z]*\)(.*)','\g<1>', value) for value in grouped_energymix['key']] 
# extracting date in new column
grouped_energymix['date'] = [re.sub('(.*)\([A-Z]*\)(.*)','\g<2>', value) for value in grouped_energymix['key']] 
```

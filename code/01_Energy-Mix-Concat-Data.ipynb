## Reading in all files and creating a big dataset for export

```python
files = glob.glob('../data/selected_countries/*.csv')
```

```python
energymix = pd.DataFrame()

for filename in files:
    df_temp = pd.read_csv(filename)
    energymix = pd.concat([energymix,df_temp],ignore_index=True)
```

```python
len(energymix)
```

```python
energymix.to_csv('../data/export-selected_countries_dataset.csv', index=False)
```
